# Transformation

### Step3: Loading and Exploring Data(Data ingestion)

```python
# Load and Explore
df= pd.read_csv("sales_data.csv")
print(df.head())
print(df.info())
print(df.describe())
```

### Step4: Data Cleaning.
-Missing values.
-Remove duplicates.
-Incorrect data types.
-inconsistent format.

```python
### Handle Misssing values
# Fill missing customer names
df['customer_name'].fillna("Unknown", inplace=True)

# Fill missing region
df['region'].fillna("Unknown", inplace=True)

# Fill missing price with median
df['price'].fillna(df['price'].median(), inplace=True)


### Fix Data Types
# Convert quantity to numeric (handle errors)
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

#### Handle Invalid Data

# Remove rows with invalid dates
df = df.dropna(subset=['order_date'])

# Remove negative or zero quantity
df = df[df['quantity'] > 0]

### Standardize Data
# Standardize region names
df['region'] = df['region'].str.title()

# Clean customer names
df['customer_name'] = df['customer_name'].str.strip()
```

### Step5: Data Transformation

```python
###Create New Columns to add meaning(Feature Engineering)
# Total sales
df['total_sales'] = df['quantity'] * df['price']

# Extract year and month
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
```

### Step 6: Data Analysis

```python

# Total sales by region
sales_by_region = df.groupby('region')['total_sales'].sum()

# Total sales by product
sales_by_product = df.groupby('product')['total_sales'].sum()

print(sales_by_region)
print(sales_by_product)

```
